```python
import random as rnd
from random import choice
```

```python
# Weapon class
class Weapon ():
    def __init__(self):
        self._launch = False
        self._hit = self._is_hit()

    def _is_hit (self):
        boolean = choice([True, False])
        return boolean

    @property
    def launch (self):
        return self._launch

    @property
    def hit (self):
        return self._hit
```

```python
# Rocket class
class Rocket (Weapon):
    def __init__(self):
        self._launch = False
        self._hit = self._is_hit()
```

```python
# PEM class
class PEM (Weapon):
    def __init__(self):
        self._launch = False
        self._hit = self._is_hit()
```
